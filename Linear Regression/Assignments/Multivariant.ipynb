```python
features = pd.DataFrame(boston.data, columns = boston.feature_names)
target = pd.DataFrame(boston.target, columns = ["TARGET"])
```

```python
data = pd.concat([features, target], axis = 1)
data
```

```python
X1 = data["RM"]
X2 = data["LSTAT"]
Y = data["TARGET"]
```

```python
X1 = (X1 - X1.min())/(X1.max() - X1.min())
X2 = (X2 - X2.min())/(X2.max() - X2.min())
Y = (Y - Y.min())/(Y.max() - Y.min())
```

```python
x1 = np.array(X1)
x2 = np.array(X2)
y = np.array(Y)
```

```python
x1_train, x1_test, x2_train, x2_test, y_train, y_test = train_test_split(x1, x2, y, test_size = 0.2)
```

```python
def hyp(a, x1, b, x2, c):
    return a*x1 + b*x2 + c
```

```python
def error(a, x1, b, x2, c, y):
    e = 0
    m = len(x1)
    
    for i in range(m):
        e += np.power(hyp(a, x1[i], b, x2[i], c) - y[i], 2)
    
    return (1/(2*m))*e
```

```python
def step_grad(a, x1, b, x2, c, y, learning_rate):
    grad_a = 0
    grad_b = 0
    grad_c = 0
    m = len(x1)
    
    for i in range(m):
        grad_a += (1/m)*(hyp(a, x1[i], b, x2[i], c) - y[i])*x1[i]
        grad_b += (1/m)*(hyp(a, x1[i], b, x2[i], c) - y[i])*x2[i]
        grad_c += (1/m)*(hyp(a, x1[i], b, x2[i], c) - y[i])
        
    a = a - grad_a*learning_rate
    b = b - grad_b*learning_rate
    c = c - grad_c*learning_rate
    
    return a, b, c
```

```python
def descnd(init_a, x1, init_b, x2, init_c, y, learning_rate, iteration):
    a = init_a
    b = init_b
    c = init_c
    
    for i in range(iteration):
        e = error(a, x1, b, x2, c, y)
        
        if i % 1000 == 0:
            print(f"Error : {np.sqrt(e)}, a : {a}, b : {b}, c : {c}")
            
        a, b, c = step_grad(a, x1, b, x2, c, y, learning_rate)
        
    return a, b, c
```

```python
a = 0.48541186766857275
b = -0.5295161607141773
c = 0.2938441369353163
learning_rate = 0.000001
iteration = 20000
```

```python
final_a, final_b, final_c = descnd(a, x1_train, b, x2_train, c, y_train, learning_rate, iteration)
```

```python
plt.plot(x1_train, y_train, 'r.',
        x1_train, hyp(a, x1_train, b, x2_train, c), 'bo',
        x1_train, hyp(final_a, x1_train, final_b, x2_train, final_c), 'g--')
```

```python
plt.plot(x2_train, y_train, 'r.',
        x2_train, hyp(a, x1_train, b, x2_train, c), 'bo',
        x2_train, hyp(final_a, x1_train, final_b, x2_train, final_c), 'g.')
```
