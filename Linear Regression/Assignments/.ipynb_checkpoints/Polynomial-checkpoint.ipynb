```python
features = pd.DataFrame(boston.data, columns = boston.feature_names) 
target = pd.DataFrame(boston.target, columns = ['TARGET'])
```

```python
X = data["RM"]
Y = data["TARGET"]
```

```python
plt.plot(X, Y, 'r.')
```

```python
X = (X - X.min())/(X.max()-X.min())
Y = (Y - Y.min())/(Y.max()-Y.min())
```

```python
x = np.array(X)
y = np.array(Y)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
def hyp(a, x, b, c):
    return a*(np.power(x,2)) + b*x + c
```

```python
def error(a, x, b, c, y):
    e = 0
    m = len(x)
    for i in range(m):
        e += np.power(hyp(a, x[i], b, c) - y[i], 2)
    return (1/(2*m))*e
```

```python
def step_grad(a, x, b, c, y, learning_rate):
    grad_a = 0
    grad_b = 0
    grad_c = 0
    m = len(x)
    
    for i in range(m):
        grad_a += (1/m)*(hyp(a, x[i], b, c) - y[i])*x[i]*x[i]
        grad_b += (1/m)*(hyp(a, x[i], b, c) - y[i])*x[i]
        grad_c += (1/m)*(hyp(a, x[i], b, c) - y[i])
        
    a = a - grad_a*learning_rate
    b = b - grad_b*learning_rate
    c = c - grad_c*learning_rate
    
    return a, b, c
    
```

```python
def descnd(init_a, x, init_b, init_c, y, learning_rate, iteration):
    a = init_a
    b = init_b
    c = init_c
    
    for i in range(iteration):
        e = error(a, x, b, c, y)
        
        if i % 1000 == 0:
            print(f"Error : {np.sqrt(e)}, a : {a}, b : {b}, c = {c}")
        a, b, c = step_grad(a, x, b, c, y, learning_rate)
        
    return a, b, c
```

```python
a = 1.5003410872052416
b = -0.6029505210903583
c = 0.2662170866008315
learning_rate = 0.001
iteration = 20000
```

```python
final_a, final_b, final_c = descnd(a, x_train, b, c, y_train, learning_rate, iteration)
```

```python
plt.plot(x_train, y_train, 'r.',
        x_train, hyp(a, x_train, b, c), 'b.',
        x_train, hyp(final_a, x_train, final_b, final_c), 'g.')
```
