```python
x = np.linspace(-1,10,1000)
```

```python
cosine = np.cos(2*np.pi*x)
```

```python
plt.plot(x, cosine)
```

```python
exp = np.e**(-x)
```

```python
plt.plot(x,exp*cosine, 'r,')
```

```python
fig = plt.figure()
plt.subplot(221)
p = plt.plot(x, cosine, 'r--')

plt.subplot(222)
plt.plot(x, exp, 'k--')

plt.subplot(212)
plt.plot(x, cosine*exp, 'y--')
```
